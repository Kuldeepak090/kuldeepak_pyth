```python
cols_list = ['bore','stroke','peak-rpm','price']
df[cols_list].info()
```

```python
bore_mean = df['bore'].astype("float").mean()
stroke_mean = df['stroke'].astype("float").mean()
peak_mean = df['peak-rpm'].astype("float").mean()
hp_mean = df['horsepower'].astype("float").mean()
price_mean = df['price'].astype("float").mean()
print(bore_mean,stroke_mean,peak_mean,hp_mean,price_mean)
```

```python
df['bore'].replace(np.nan,bore_mean,inplace=True)
df['bore']= df['bore'].astype(float)
```

```python
df['peak-rpm'].replace(np.nan,bore_mean,inplace=True)
df['peak-rpm']= df['peak-rpm'].astype(float)
df['stroke'].replace(np.nan,bore_mean,inplace=True)
df['storke']= df['stroke'].astype(float)
df['horsepower'].replace(np.nan,bore_mean,inplace=True)
df['horsepower']= df['horsepower'].astype(float)
df['price'].replace(np.nan,bore_mean,inplace=True)
df['price']= df['price'].astype(float)


```
